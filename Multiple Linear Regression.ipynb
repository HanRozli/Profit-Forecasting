```python
# Multiple Linear Regression 
# Multivariate Linear Regression

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# importing dataset

data_set = pd.read_csv('https://trello-attachments.s3.amazonaws.com/60596dc01945760f799067c1/607e52cfbf5d601da434fa03/e6a083c192d3f7d9aa73f9bd879e408b/Startups_Data.csv')
data_set.head()
```

```python
# Extract Independent And Dependent Variable
x = data_set.iloc[:,:-1].values
y = data_set.iloc[:,4].values
```

```python
# Encoding Dummy Variable
# Use LabelEncoder OneHotEncoding ColumnTransformer

from sklearn.preprocessing import LabelEncoder , OneHotEncoder
from sklearn.compose import ColumnTransformer

lb = LabelEncoder()

x[:,3] = lb.fit_transform(x[:,3])

# x [California --> 0   Florida --> 1   New York --> 2]

transformer = ColumnTransformer(transformers=[(
    "OneHot",
    OneHotEncoder(),
    [3])], remainder='passthrough')

x = transformer.fit_transform(x.tolist())
x = x.astype('int')

# remainder = 'passthrough' DO NOT APPLY ANYTHING ON REMAINING COLUMN
```
